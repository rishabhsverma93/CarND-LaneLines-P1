## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
# points for area of interest
top_left = [400, 330]
bottom_left = [50, 540]
top_right = [600, 330]
bottom_right = [900, 540]
#contours = np.empty([2])
#np.append([contours], [bottom_left, top_left, top_right, bottom_right], axis=0)
````

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

# Global variables taken to calculate average
left_top_list = []
left_bottom_list = []
right_top_list = []
right_bottom_list = []

def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)
    """
    # left_lines and right lines arrays to store lines with different slopes. also contains y top and bottom for left and right
    left_lines = []
    right_lines = []
    left_bottom_point = bottom_left
    left_top_point = top_left
    right_bottom_point = bottom_right
    right_top_point = top_right
    left_lines_y_bottom = top_left[1]
    left_lines_y_top = bottom_left[1]
    right_lines_y_bottom = top_right[1]
    right_lines_y_top = bottom_right[1]
    
    """
    Image:
    We go over each line and find out the highest possible and lowest possible y coordinates within the 
    region of interest from all lines on both left and right sides. This way, we are able to find left
    and right bottom and top points. in the case of an image, this would suffice
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            # if slope of line > 0, then line is part of left line. otherwise, right line
            if((y2-y1)/(x2-x1) >= 0):
                left_lines.append(line)
                if(y1 <= left_lines_y_top):
                    left_lines_y_top = y1
                    left_top_point = (x1, y1)
                if(y2 <= left_lines_y_top):
                    left_lines_y_top = y2
                    left_top_point = (x2, y2)
                if(y1 >= left_lines_y_bottom):
                    left_lines_y_bottom = y1
                    left_bottom_point = (x1, y1)
                if(y2 >= left_lines_y_bottom):
                    left_lines_y_bottom = y2
                    left_bottom_point = (x2, y2)
            else:
                right_lines.append(line)
                if(y1 <= right_lines_y_top):
                    right_lines_y_top = y1
                    right_top_point = (x1, y1)
                if(y2 <= right_lines_y_top):
                    right_lines_y_top = y2
                    right_top_point = (x2, y2)
                if(y1 >= right_lines_y_bottom):
                    right_lines_y_bottom = y1
                    right_bottom_point = (x1, y1)
                if(y2 >= left_lines_y_bottom):
                    right_lines_y_bottom = y2
                    right_bottom_point = (x2, y2)
    """
    Video:
    In case of video however, simply plotting these lines gives rise to too much flickering. Therefore we average
    a fixed number of points (equal to sliding window size or however many are available till a given frame 
    in the video) and then plot the lines according to that
    """
    # Adding the points in the list (the list grows bigger with increasing number of frames)
    left_top_list.append(left_top_point)
    left_bottom_list.append(left_bottom_point)
    right_top_list.append(right_top_point)
    right_bottom_list.append(right_bottom_point)
    
    # Calculating the sliding window size
    sliding_window_default_size = 50
    sliding_window_size = 0
    if(len(left_top_list) < sliding_window_default_size):
        sliding_window_size = len(left_top_list)
    else:
        sliding_window_size = sliding_window_default_size
    
    # We sum all the tuples in left_top_list and get one tuple, which is the sum of the most recent points (for 
    # which the number is equal to the size of the sliding window 
    sum_left_top = tuple([sum(x) for x in zip(*left_top_list[-(sliding_window_size):])])
    # We then divide both x and y coordinates of the tuple sum to find the averaged top_left point
    avg_left_top = tuple(ti/sliding_window_size for ti in sum_left_top)
    
    # We do the same for the other 3 sets of points : left_bottom
    sum_left_bottom = tuple([sum(x) for x in zip(*left_bottom_list[-(sliding_window_size):])])
    avg_left_bottom = tuple(ti/sliding_window_size for ti in sum_left_bottom)
    # Right_top
    sum_right_top = tuple([sum(x) for x in zip(*right_top_list[-(sliding_window_size):])])
    avg_right_top = tuple(ti/sliding_window_size for ti in sum_right_top)
    # Right_bottom
    sum_right_bottom = tuple([sum(x) for x in zip(*right_bottom_list[-(sliding_window_size):])])
    avg_right_bottom = tuple(ti/sliding_window_size for ti in sum_right_bottom)
    
    # We finally plot the lines using the averaged points
    cv2.line(img, tuple(int(x) for x in avg_left_top), tuple(int(x) for x in avg_left_bottom), color, 10)
    cv2.line(img, tuple(int(x) for x in avg_right_top), tuple(int(x) for x in avg_right_bottom), color, 10)
    
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````
